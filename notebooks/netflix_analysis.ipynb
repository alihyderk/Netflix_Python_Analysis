```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/netflix_titles.csv")
df.head()

df.info()

df.shape
df.isnull().sum()
df.duplicated().sum()

df['type'].value_counts()
df['country'].value_counts().head(10)

df['date_added'] = df['date_added'].str.strip()        # remove extra spaces
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce')

df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index()

# check missing values
df.isnull().sum()

# clean missing directors and cast by filling with "Unknown"
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")

# fill missing country with "Unknown"
df['country'] = df['country'].fillna("Unknown")

df.isnull().sum()

year_counts = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=year_counts.index, y=year_counts.values)

plt.title("Netflix Titles Added Over the Years", fontsize=16)
plt.xlabel("Year Added to Netflix")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='type', palette='viridis')

plt.title("Count of Movies vs TV Shows", fontsize=16)
plt.xlabel("Type of Content")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

df_country = df['country'].dropna().str.split(',', expand=True).stack().str.strip()

top_countries = df_country.value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='mako')

plt.title("Top 10 Countries Producing Netflix Content", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()


# Split multiple countries and clean the data
df_country = df['country'].dropna().str.split(',', expand=True).stack().str.strip()

# Remove Unknown or blank entries
df_country = df_country[df_country != "Unknown"]
df_country = df_country[df_country != ""]     # remove empty strings if any

# Get top 10 countries
top_countries = df_country.value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='mako')

plt.title("Top 10 Countries Producing Netflix Content (Cleaned)", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()

plt.figure(figsize=(10,6))
sns.countplot(y=df['rating'], order=df['rating'].value_counts().index, palette="viridis")

plt.title("Distribution of Content Ratings on Netflix", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Rating")
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()

plt.figure(figsize=(8,6))

sns.countplot(x=df['type'], palette="Set2")

plt.title("Movies vs TV Shows on Netflix", fontsize=16)
plt.xlabel("Type of Content")
plt.ylabel("Number of Titles")
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()

# Split the listed_in column and explode into separate rows
genres = df['listed_in'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
genres.name = 'genre'

# Count genre frequency
genre_counts = genres.value_counts().head(15)

# Plot the top 15 genres
plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="mako")

plt.title("Top 15 Most Frequent Genres on Netflix", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()

# Remove empty and unknown directors
directors = df['director'].replace("Unknown", np.nan).dropna()

# Split directors into separate names
directors = directors.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
directors.name = 'director_name'

# Count top 15 directors
top_directors = directors.value_counts().head(15)

# Plot the chart
plt.figure(figsize=(12,6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="viridis")

plt.title("Top 15 Directors with the Most Titles on Netflix", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()

# Top 20 Most Frequent Actors

import numpy as np

# Remove empty or missing cast entries
cast = df['cast'].replace("Unknown", np.nan).dropna()

# Split multiple actors and convert to a long list
cast = cast.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
cast.name = 'actor_name'

# Count top 20 actors
top_actors = cast.value_counts().head(20)

# Plot the chart
plt.figure(figsize=(12,6))
sns.barplot(x=top_actors.values, y=top_actors.index, palette="viridis")

plt.title("Top 20 Actors Appearing Most Frequently on Netflix", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Actor")
plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.show()

# Movie Duration Distribution

# Filter only movies
movies = df[df['type'] == 'Movie']

# Extract numeric duration (remove " min")
movies['duration_minutes'] = movies['duration'].str.replace(" min", "", regex=False)

# Convert to numeric
movies['duration_minutes'] = pd.to_numeric(movies['duration_minutes'], errors='coerce')

# Drop missing durations
movies = movies.dropna(subset=['duration_minutes'])

# Plot distribution
plt.figure(figsize=(12,6))
sns.histplot(movies['duration_minutes'], bins=40, kde=True, color='teal')

plt.title("Distribution of Movie Durations on Netflix", fontsize=16)
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()

# TV Show Seasons Analysis

# Filter only TV Shows
tv_shows = df[df['type'] == 'TV Show']

# Extract numeric seasons (remove " Seasons" or " Season")
tv_shows['seasons'] = tv_shows['duration'].str.replace(" Seasons", "", regex=False)
tv_shows['seasons'] = tv_shows['seasons'].str.replace(" Season", "", regex=False)

# Convert to numeric
tv_shows['seasons'] = pd.to_numeric(tv_shows['seasons'], errors='coerce')

# Drop missing values
tv_shows = tv_shows.dropna(subset=['seasons'])

# Plot
plt.figure(figsize=(12,6))
sns.countplot(x='seasons', data=tv_shows, palette='viridis')

plt.title("Distribution of Number of Seasons in Netflix TV Shows", fontsize=16)
plt.xlabel("Number of Seasons")
plt.ylabel("Number of TV Shows")
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()

# Movies vs TV Shows Added Over the Years

# Filter rows with valid dates
df_clean = df.dropna(subset=['date_added'])

# Extract year
df_clean['year_added'] = df_clean['date_added'].dt.year

# Group data
movies_per_year = df_clean[df_clean['type'] == 'Movie']['year_added'].value_counts().sort_index()
shows_per_year = df_clean[df_clean['type'] == 'TV Show']['year_added'].value_counts().sort_index()

# Plot
plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index, movies_per_year.values, label='Movies Added', marker='o')
plt.plot(shows_per_year.index, shows_per_year.values, label='TV Shows Added', marker='o')

plt.title("Movies vs TV Shows Added on Netflix Over the Years", fontsize=16)
plt.xlabel("Year Added")
plt.ylabel("Number of Titles Added")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)

plt.show()




















```

```python
from wordcloud import WordCloud, STOPWORDS

# Combine all descriptions into one big string
text = " ".join(df['description'].dropna().tolist())

# Extra stopwords to remove boring words
stopwords = set(STOPWORDS)
stopwords.update(['series', 'season', 'film', 'story', 'one'])

# Create word cloud
wordcloud = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    stopwords=stopwords,
    colormap='viridis'
).generate(text)

# Plot it
plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Netflix Descriptions", fontsize=18)
plt.show()
```
